```python
import numpy as np
import pandas as pd
import json
import os
```

```python
TRANSCRIPT_DIR="transcript"
AUDIO_DIR="audio"
```

```python
df=pd.read_csv("data.csv")
df
```

```python
def gettranscript(ids,lang):
    return os.path.join(TRANSCRIPT_DIR,ids+'.'+lang+'.json')
```

```python
def getaudio(ids,lang):
    return os.path.join(AUDIO_DIR,ids+'.'+lang+'.wav')
```

```python
df_trans_hi=[]
for i in df['File']:
    file1 = open(gettranscript(i,'hi'),encoding='utf-8')
    data_hi=json.load(file1)
    df_hi=pd.DataFrame.from_dict(data_hi)
    df_hi['audio']=i
    df_trans_hi.append(df_hi)
```

```python
df_hi = pd.concat(df_trans_hi,ignore_index=True)
```

```python
df_hi.columns=['text_hi', 'start_hi', 'duration_hi', 'audio']
```

```python
df_hi['text_hi'] = df_hi['text_hi'].str.replace(' -', '')
```

```python
df_hi['text_hi'] = df_hi['text_hi'].str.replace('- ', '')
```

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/19OAPQaspQKl1gyTKZbmzEimFps4Ppj-vclR3p2T3EXk/edit#gid=319756618'
csv_export_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df_en = pd.read_csv(csv_export_url)
```

```python
df_en['text_hi'] = df_en['text_hi'].str.replace('- ', '')
```

```python
df_en['text_hi'] = df_en['text_hi'].str.replace(' -', '')
```

```python
dfinal = df_en.merge(df_hi, how='inner', on=['text_hi','audio'])
```

```python
dfinal
```

```python
dfinal = dfinal.drop_duplicates(subset=['text_hi', 'audio'])
```

```python
dfinal
```

```python
backup_df=dfinal
```

```python
AUDIO_PATH="dataset"
if not os.path.exists(AUDIO_PATH):
    os.makedirs(AUDIO_PATH)
    
def getaudio1(ids,lang):
    return os.path.join(AUDIO_PATH,ids+'.'+lang+'.wav')
```

```python
from pydub import AudioSegment

for i in dfinal.index:
    audio_en = AudioSegment.from_wav(getaudio(dfinal.loc[i, 'audio'], 'en'))
    audio_hi = AudioSegment.from_wav(getaudio(dfinal.loc[i, 'audio'], 'hi'))

    start_time_en = max(dfinal.loc[i, 'start'] * 1000 - 500,0) # Convert to milliseconds
    end_time_en = start_time_en + dfinal.loc[i, 'duration'] * 1000 + 500  # Calculate end time

    start_time_hi = max(dfinal.loc[i, 'start_hi'] * 1000 - 500,0)  # Convert to milliseconds
    end_time_hi = start_time_hi + dfinal.loc[i, 'duration_hi'] * 1000 + 500  # Calculate end time

    end_time_en = min(end_time_en, len(audio_en))
    end_time_hi = min(end_time_hi, len(audio_hi))
    cut_audio_en = audio_en[start_time_en:end_time_en]
    cut_audio_hi = audio_hi[start_time_hi:end_time_hi]
    
    output_audio_name = dfinal.loc[i, 'audio'] + '_' + str(i)
    dfinal.loc[i,'id']=output_audio_name
    cut_audio_en.export(getaudio1(output_audio_name, 'en'), format="wav")
    cut_audio_hi.export(getaudio1(output_audio_name, 'hi'), format="wav")

    print(f'Progress: {i+1}/{len(dfinal)}')
```

```python
dfinal=dfinal[['id','text','text_hi']]
```
