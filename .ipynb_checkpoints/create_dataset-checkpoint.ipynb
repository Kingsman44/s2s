```python
import numpy as np
import pandas as pd
import json
import os
```

```python
TRANSCRIPT_DIR="transcript"
AUDIO_DIR="audio"
```

```python
df=pd.read_csv("data.csv")
df
```

```python
def gettranscript(ids,lang):
    return os.path.join(TRANSCRIPT_DIR,ids+'.'+lang+'.json')
```

```python
def getaudio(ids,lang):
    return os.path.join(AUDIO_DIR,ids+'.'+lang+'.wav')
```

```python
df_trans_en=[]
df_trans_hi=[]
for i in df['File']:
    file1 = open(gettranscript(i,'hi'),encoding='utf-8')
    data_hi=json.load(file1)
    file = open(gettranscript(i,'en'),encoding='utf-8')
    data_en=json.load(file)
    df_hi=pd.DataFrame.from_dict(data_hi)
    df_en=pd.DataFrame.from_dict(data_en)
    df_en['audio']=i
    df_hi['audio']=i
    df_trans_en.append(df_en)
    df_trans_hi.append(df_hi)
```

```python
df_trans_en
```

```python
df_en = pd.concat(df_trans_en,ignore_index=True)
df_hi = pd.concat(df_trans_hi,ignore_index=True)
```

```python
df_en['text_hi']=df_en['text']
```

```python
df_hi
```

```python
df_hi.columns=['text_hi', 'start_hi', 'duration_hi', 'audio']
```

```python
df_hi
```

```python
df_en
```

```python
from deep_translator import GoogleTranslator
trans=GoogleTranslator(source='en', target='hi')
```

```python
for i in df_en.index:
    df_en.loc[i,'text_hi']=trans.translate(df_en.loc[i,'text'])
    print(f'Progress: {i}/{len(df_en.index)}')
```

```python
from deep_translator import GoogleTranslator
import pandas as pd
import multiprocessing

trans = GoogleTranslator(source='en', target='hi')
```

```python
def translate_text(row):
    row['text_hi'] = trans.translate(row['text'])
    return row

# Number of parallel processes (you can adjust this based on your CPU cores)
num_processes = multiprocessing.cpu_count()

# Split the DataFrame into chunks for parallel processing
chunks = [df_en.loc[i:i + num_processes - 1] for i in range(0, len(df_en), num_processes)]

# Function to perform translation in parallel
def parallel_translation(chunk):
    translated_chunk = chunk.apply(translate_text, axis=1)
    return translated_chunk

# Create a pool of worker processes and perform translation in parallel
with multiprocessing.Pool(processes=num_processes) as pool:
    results = pool.map(parallel_translation, chunks)

# Concatenate the results back into a single DataFrame
df_result = pd.concat(results)

print(df_result)
```

```python
df_use = pd.merge(df_en, df_hi, on=['text_hi', 'audio'], how='inner')
```

```python
df_use
```

```python
AUDIO_PATH="dataset"
if not os.path.exists(AUDIO_PATH):
    os.makedirs(AUDIO_PATH)
    
def getaudio1(ids,lang):
    return os.path.join(AUDIO_PATH,ids+'.'+lang+'.wav')
```

```python
from pydub import AudioSegment

for i in df_use.index:
    # Load the WAV file
    audio_en = AudioSegment.from_wav(getaudio(df_use.loc[i,'audio'],'en'))
    audio_hi = AudioSegment.from_wav(getaudio(df_use.loc[i,'audio'],'hi'))
    cut_audio_en = audio_en[df_use.loc[i,'start']:df_use.loc[i,'start'] + df_use.loc[i,'duration']]
    cut_audio_hi = audio_hi[df_use.loc[i,'start_hi']:df_use.loc[i,'start_hi'] + df_use.loc[i,'duration_hi']]
    df_use.loc[i,'audio']=df_use.loc[i,'audio']+'_'+str(i)
    # Export the cut audio to a new WAV file
    cut_audio_en.export(getaudio1(df_use.loc[i,'audio'],'en'), format="wav")
    cut_audio_hi.export(getaudio1(df_use.loc[i,'audio'],'hi'), format="wav")
    print(f'Progress: {i}/{len(df_use.index)}')
```

```python
from pydub import AudioSegment

for i in df_use.index:
    audio_en = AudioSegment.from_wav(getaudio(df_use.loc[i, 'audio'], 'en'))
    audio_hi = AudioSegment.from_wav(getaudio(df_use.loc[i, 'audio'], 'hi'))

    start_time_en = max(df_use.loc[i, 'start'] * 1000 - 500,0) # Convert to milliseconds
    end_time_en = start_time_en + df_use.loc[i, 'duration'] * 1000 + 500  # Calculate end time

    start_time_hi = max(df_use.loc[i, 'start_hi'] * 1000 - 500,0)  # Convert to milliseconds
    end_time_hi = start_time_hi + df_use.loc[i, 'duration_hi'] * 1000 + 500  # Calculate end time

    end_time_en = min(end_time_en, len(audio_en))
    end_time_hi = min(end_time_hi, len(audio_hi))
    cut_audio_en = audio_en[start_time_en:end_time_en]
    cut_audio_hi = audio_hi[start_time_hi:end_time_hi]
    
    output_audio_name = df_use.loc[i, 'audio'] + '_' + str(i)
    df_use.loc[i,'id']=output_audio_name
    cut_audio_en.export(getaudio1(output_audio_name, 'en'), format="wav")
    cut_audio_hi.export(getaudio1(output_audio_name, 'hi'), format="wav")

    print(f'Progress: {i+1}/{len(df_use)}')
```
