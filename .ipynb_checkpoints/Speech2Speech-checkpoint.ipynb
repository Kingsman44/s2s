```python
import pandas as pd
import numpy as np
import json
import seaborn as sns
import matplotlib.pyplot as plt
import os
import librosa
import librosa.display
import IPython.display as ipd
```

```python
DATASET_DIR="dataset"
LANGS=['en','hi']
SR=16000
```

```python
df=pd.read_csv('final_dataset.csv')
df.columns
```

```python
df=df[['id', 'text', 'text_hi']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
def getaudiofile(ids,lang):
    return os.path.join(DATASET_DIR,ids+'.'+lang+'.wav')
```

```python
AUDIO_PATH="dataset_denoise"
if not os.path.exists(AUDIO_PATH):
    os.makedirs(AUDIO_PATH)
    
def getaudio1(ids,lang):
    return os.path.join(AUDIO_PATH,ids+'.'+lang+'.wav')
```

```python
from denoise import AudioDeNoise

for ids in df['id']:
    if not os.path.exists(getaudio1(ids,'en')):
        audioDenoiser = AudioDeNoise(inputFile=getaudiofile(ids,'en'))
        audioDenoiser.deNoise(outputFile=getaudio1(ids,'en'))    
    if not os.path.exists(getaudio1(ids,'hi')):
        audioDenoiser = AudioDeNoise(inputFile=getaudiofile(ids,'hi'))
        audioDenoiser.deNoise(outputFile=getaudio1(ids,'hi'))           
    
```

```python
print("English Voice")
ipd.Audio(getaudiofile(df['id'][1],'en'),rate=SR)
```

```python
print("English Voice After denoise")
ipd.Audio(getaudio1(df['id'][1],'en'),rate=SR)
```

```python
print("Hindi Voice")
ipd.Audio(getaudiofile(df['id'][1],'hi'),rate=SR)
```

```python
print("Hindi Voice After Denoise")
ipd.Audio(getaudio1(df['id'][1],'hi'),rate=SR)
```

```python
y,sr=librosa.load(getaudio1(df['id'][1],'en'),sr=SR)
sr
```

```python
en_audio=[]
hi_audio=[]
for i in df.index:
    audio_data_en, _ = librosa.load(getaudio1(df['id'][i], 'en'),sr=SR)
    audio_data_hi, _ = librosa.load(getaudio1(df['id'][i], 'hi'),sr=SR)
    en_audio.append(audio_data_en)
    hi_audio.append(audio_data_hi)
```

```python
new_dict={
    'en_audio':en_audio,
    'en_text':df['text'],
    'hi_text':df['text_hi'],
    'hi_audio':hi_audio
}
```

```python
new_df=pd.DataFrame.from_dict(new_dict)
```

```python
new_df
```

```python
pd.Series(new_df['en_audio'][0]).plot(figsize=(10, 5),
                  lw=1,
                  title='English Audio')
plt.show()
```

```python
pd.Series(new_df['hi_audio'][0]).plot(figsize=(10, 5),
                  lw=1,
                  title='Hindi Audio')
plt.show()
```

```python
audio_data_en = new_df['en_audio'][0]
D = librosa.stft(audio_data_en)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db, x_axis='time', y_axis='log', ax=ax)
ax.set_title('Spectrogram English Audio', fontsize=20)
fig.colorbar(img, ax=ax, format='%0.2f')  # Fix typo in colorbar function name
plt.show()
```

```python
audio_data_hi = new_df['hi_audio'][0]
D = librosa.stft(audio_data_hi)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db, x_axis='time', y_axis='log', ax=ax)
ax.set_title('Spectrogram Hindi Audio', fontsize=20)
fig.colorbar(img, ax=ax, format='%0.2f')  # Fix typo in colorbar function name
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=audio_data_en,
                                   sr=SR,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = librosa.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram English Audio', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=audio_data_hi,
                                   sr=SR,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = librosa.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram Hindi Audio', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
# Loading model
from transformers import AutoFeatureExtractor
model = "ntu-spml/distilhubert"

feature_extractor = AutoFeatureExtractor.from_pretrained(model, do_normalize = True, return_attention_mask = True)
```

```python
sample = new_df['en_audio'][0]
print('Example Data:')
print(f"Mean: {np.mean(sample):.3}")
print(f"Variance: {np.var(sample):.3}")
```

```python
inputs = feature_extractor(sample, sampling_rate = SR)
print(f"\nInputs Keys: {list(inputs.keys())}\n")
print(f" Mean: {np.mean(inputs['input_values']):.3}, Variance: {np.var(inputs['input_values']):.3}")
```

```python
new_df['norm_en'] = [feature_extractor(audio_data, sampling_rate = SR) for audio_data in new_df['en_audio']]
new_df['norm_hi'] = [feature_extractor(audio_data, sampling_rate = SR) for audio_data in new_df['hi_audio']]
```

```python
new_df['norm_en'][0]['attention_mask'][0]
```

```python
def getinput(data,ip):
    return data[ip][0]

new_df['norm_audio_en'] = [getinput(audio_data,'input_values') for audio_data in new_df['norm_en']]
new_df['norm_audio_hi'] = [getinput(audio_data,'input_values') for audio_data in new_df['norm_hi']]

new_df['attention_mask_en'] = [getinput(audio_data,'attention_mask') for audio_data in new_df['norm_en']]
new_df['attention_mask_hi'] = [getinput(audio_data,'attention_mask') for audio_data in new_df['norm_hi']]
```

```python
new_df
```

```python
pd.Series(new_df['norm_audio_en'][0]).plot(figsize=(10, 5),
                  lw=1,
                  title='Normalised English Audio')
plt.show()
```

```python
pd.Series(new_df['norm_audio_en'][0]).plot(figsize=(10, 5),
                  lw=1,
                  title='Normalised Hindi Audio')
plt.show()
```

```python
def extract_mfcc(audio_data):
    mfcc = librosa.feature.mfcc(y=audio_data, sr=SR, n_mfcc=13)
    return mfcc.T

new_df['en_mfcc'] = [extract_mfcc(audio_data) for audio_data in new_df['norm_audio_en']]
new_df['hi_mfcc'] = [extract_mfcc(audio_data) for audio_data in new_df['norm_audio_hi']]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(new_df['en_text'].values.tolist() + new_df['hi_text'].values.tolist())

new_df['en_texts'] = tokenizer.texts_to_sequences(new_df['en_text'].values)
new_df['hi_texts'] = tokenizer.texts_to_sequences(new_df['hi_text'].values)
```
